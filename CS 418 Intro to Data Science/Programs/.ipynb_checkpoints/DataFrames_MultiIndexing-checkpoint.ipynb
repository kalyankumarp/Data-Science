```python
import pandas as pd
from numpy.random import randn
```

```python
outside0 = ["F1","F1","F1","F1","F1","F1","F1","F1","F1", "F2","F2","F2","F2","F2","F2","F2","F2","F2", "F3","F3","F3","F3","F3","F3","F3","F3","F3"]
outside1 = ["G1","G1","G1","G2","G2","G2","G3","G3","G3","G1","G1","G1","G2","G2","G2","G3","G3","G3","G1","G1","G1","G2","G2","G2","G3","G3","G3"]
inside = [1,2,3,1,2,3,1,2,3,1,2,3,1,2,3,1,2,3,1,2,3,1,2,3,1,2,3]
hier_index = list(zip(outside0, outside1, inside))
hier = pd.MultiIndex.from_tuples(hier_index)
```
